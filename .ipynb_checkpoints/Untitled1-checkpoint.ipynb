```python
class Room:
    def __init__(self):
        self.length = 12
        self.breadth = 24  # Corrected from self.length = 24

    def area(self):
        # Corrected typo: self.breadth instead of self.bredth
        print(self.length * self.breadth)

hall = Room()
hall.area()
```

```python
class done:
    
    def __init__(self):
        self.check = "not done"
        
    def is_done(self):
        self.check="done"
        print(self.check)

c = done()
c.is_done()
```

```python
class College:
    def __init__(self, branch):
        self.branch = branch 

    def pr(self):
        print(self.branch)

class Student(College):
    def __init__(self, branch, usn):
        super().__init__(branch)
        self.usn = usn

my_student = Student("ECE", "1rv24ec054")
my_student.pr()         
print(my_student.usn)   
```

```python
class vahan:
    def info(self):
        print("in main class")

class car(vahan):
    def __init__(self):
        super().info()
    def info(self):
        print("in sub class")

vehicle = car()
vehicle.info()
```

```python
class India:
    def capital(self):
        print("dehli")
    def lang(self):
        print("no official lang")
    def type(self):
        print("developing")

class USA:
    def capital(self):
        print("washington")
    def lang(self):
        print("English")
    def type(self):
        print("developed")

obj = []
i=9
while(1>=0):
    if i%2 == 0:
        obj[i]=India();
    else:
        obj[i]=USA();
    i-=i

for country in (obj[]):
    country.capital()
    country.lang()
    country.type()
```
