# Stack

```python
#Stack
class stack:
    top = -1
    arr = []
    #def __init__(self):
       
    def push(top):
        top += 1
        arr.append(top)
        #arr.append(a)  

    def pop():
        top -= 1
        print(arr[top])

    def main():
        obj = stack()
        obj.push(10)
        obj.pop()

if "__name__" == "__main__":
    main()
```

```python
def check(str):
    open = "[{("
    close = ")}]"
    stack = []
    for char in str:
        if char in open:
            stack.append(char)
        elif char in close:
            if len(stack) == 0:
                return False
            if open.index(stack.pop()) != close.index(char):
                return False
    return len(stack) == 0
```

# Queues

```python
n = int(input("people"))
k = int(input("Skipped"))

def josef_seq(n,k):
    s = []
    q = []
    for i in range (1,n+1):
        q.append(i)
    j=1

    while len(q)>1:
        item = q.pop()
        if j%k == 0:
            s.append(item)
        else:
            q.append(item)
    return s

print(josef_seq(n,k))
```
