# Bubble Sort

```python
#Bubble sort
def bubble_sort(S):
    for j in range(len(S)-1):
        for i in range(len(S)-j-1):
            if(S[i] >= S[i+1]):
                S[i], S[i + 1] = S[i + 1], S[i]

S = [10,20,30,1,5,8,2,36]
bubble_sort(S)
print(S)
```

# Selection Sort

```python
#Sel sort
def selection_sort(S):
    n = len(S)
    for j in range(n - 1):
        min_idx = j
        for i in range(j + 1, n):
            if S[i] < S[min_idx]:
                min_idx = i
        S[j], S[min_idx] = S[min_idx], S[j]

S = [10,20,30,1,5,8,2,36]
selection_sort(S)
print(S) 
```

#   Insertion Sort

```python
#insertion sort
def insertionSort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

S = [10,20,30,1,5,8,2,36]
insertionSort(S)
print(S) 
```
