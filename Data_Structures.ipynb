```python
def check(str):
    open = "[{("
    close = ")}]"
    stack = []
    for char in str:
        if char in open:
            stack.append(char)
        elif char in close:
            if len(stack) == 0:
                return False
            if open.index(stack.pop()) != close.index(char):
                return False
    return len(stack) == 0
```
