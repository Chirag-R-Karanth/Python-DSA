```python
def josef_seq(n, k):
    s = []
    # Create the circle of people
    q = [i for i in range(1, n + 1)]
    j = 1

    while len(q) > 0:
        # Pop from the FRONT to maintain order
        item = q.pop(0) 
        if j % k == 0:
            s.append(item)
        else:
            # Put them back at the end of the line
            q.append(item)
        j += 1
    return s

n = int(input("People: "))
k = int(input("Skipped: "))
print(josef_seq(n, k))
```

```python
class Queue:
    def __init__(self):
        self.queue = []

    def is_empty(self):
        return True if len(self.queue) == 0 else False
    
    def enqueue(self,item):
        self.queue.append(item)
        print(self.queue)

    def dequeue(self):
        return None if self.is_empty() else self.queue.pop(0)
    
```

```python
from collections import deque

queue = deque()
# Enqueue items
queue.append("User 1")
queue.append("User 2")
queue.append("User 3")

print(f"Current Queue: {list(queue)}") # ['User 1', 'User 2', 'User 3']

# Dequeue item (FIFO)
first_user = queue.popleft() 
print(f"Served: {first_user}")         # 'User 1'
print(f"Remaining: {list(queue)}")     # ['User 2', 'User 3']
```

# Sequencial Search

```python
from random import randint

def do_exp(floor,breaking):
    return floor >= breaking

def find_height(height,breaking):
    for i in range (1,height+1):
        if do_exp(i,breaking):
            return i-1
    return height

height = int(input("Input num of floors:"))
breaking = randint(1,height)
floor = find_height(height,breaking)
print("your egg is safe till {}".format(floor))
```

```python
def word_count(inp,search):
    len(inp)
    count = 0
    for i in inp:
        if(i == search):
            count+= 1
    return count

```

```python
def roman_to_binary(s: str) -> str:
    # 1. Hash function (Dictionary mapping)
    map = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
    
    # 2. Calculate decimal value
    total = 0
    for i in range(len(s)):
        # If current value < next value, subtract it
        if i + 1 < len(s) and map[s[i]] < map[s[i+1]]:
            total -= map[s[i]]
        else:
            total += map[s[i]]
            
    # 3. Convert decimal to binary
    print(total)
    return bin(total)[2:]

# Example: "IX" -> 9 -> "1001"
print(roman_to_binary("MCXI")) 
```
