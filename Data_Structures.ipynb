```python
def check(str):
    open = "[{("
    close = ")}]"
    stack = []
    for char in str:
        if char in open:
            stack.append(char)
        elif char in close:
            if len(stack) == 0:
                return False
            if open.index(stack.pop()) != close.index(char):
                return False
    return len(stack) == 0
```

# Queues

```python
def josef_seq(n, k):
    s = []
    # Create the circle of people
    q = [i for i in range(1, n + 1)]
    j = 1

    while len(q) > 0:
        # Pop from the FRONT to maintain order
        item = q.pop(0)
        if j % k == 0:
            s.append(item)
        else:
            # Put them back at the end of the line
            q.append(item)
        j += 1
    return s

n = int(input("People: "))
k = int(input("Skipped: "))
print(josef_seq(n, k))
```

```python
class Queue:
    def __init__(self):
        self.queue = []

    def is_empty(self):
        return True if len(self.queue) == 0 else False

    def enqueue(self,item):
        self.queue.append(item)
        print(self.queue)

    def dequeue(self):
        return None if self.is_empty() else self.queue.pop(0)

```

```python
Q = Queue()
#Q.queue = []
Q.is_empty()
Q.enqueue(10)
Q.enqueue(20)
Q.enqueue(30)
Q.dequeue()
Q.dequeue()
Q.dequeue()
print(Q.dequeue())
```
