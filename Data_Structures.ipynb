```python
def josef_seq(n, k):
    s = []
    # Create the circle of people
    q = [i for i in range(1, n + 1)]
    j = 1

    while len(q) > 0:
        # Pop from the FRONT to maintain order
        item = q.pop(0)
        if j % k == 0:
            s.append(item)
        else:
            # Put them back at the end of the line
            q.append(item)
        j += 1
    return s

n = int(input("People: "))
k = int(input("Skipped: "))
print(josef_seq(n, k))
```

```python
class Queue:
    def __init__(self):
        self.queue = []

    def is_empty(self):
        return True if len(self.queue) == 0 else False

    def enqueue(self,item):
        self.queue.append(item)
        print(self.queue)

    def dequeue(self):
        return None if self.is_empty() else self.queue.pop(0)

```

```python
from random import randint

def do_exp(floor,breaking):
    return floor >= breaking

def find_height(height,breaking):
    for i in range (1,height+1):
        if do_exp(i,breaking):
            return i-1
    return height

height = int(input("Input num of floors:"))
breaking = randint(1,height)
floor = find_height(height,breaking)
print("your egg is safe till {}".format(floor))
```

```python
def roman_to_binary(s: str) -> str:
    # 1. Hash function (Dictionary mapping)
    map = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}

    # 2. Calculate decimal value
    total = 0
    for i in range(len(s)):
        # If current value < next value, subtract it
        if i + 1 < len(s) and map[s[i]] < map[s[i+1]]:
            total -= map[s[i]]
        else:
            total += map[s[i]]

    # 3. Convert decimal to binary
    print(total)
    return bin(total)[2:]

# Example: "IX" -> 9 -> "1001"
print(roman_to_binary("MCXI"))
```

*   Bubble Sort

```python
arr = [1, 3, 2, 6, 5, 9, 8, 7]

def bubble_sort(arr):
    n = len(arr)
    for _ in range(n):
        for j in range(0, n - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

bubble_sort(arr)
print(arr)
```

```python
#Swapping 2 values
def swap(S,x,y):
    t = S[x]
    S[x] = S[y]
    S[y] = t

S = [10,20,30]
print(S)
swap(S,1,2)
print(S)
```

* Selection Sort

```python
def selection_sort(S):
    n = len(S)
    for j in range(n - 1):
        min_idx = j
        for i in range(j + 1, n):
            if S[i] < S[min_idx]:
                min_idx = i
        S[j], S[min_idx] = S[min_idx], S[j]

S = [10,20,30,1,5,8,2,36]
selection_sort(S)
print(S)
```

*   Insertion Sort

```python
def insertionSort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

S = [10,20,30,1,5,8,2,36]
insertionSort(S)
print(S)
```

* Quick sort

```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)

# Example usage:
S = [10, 20, 30, 1, 5, 8, 2, 36]
print(f"Original list: {S}")
sorted_S = quicksort(S)
print(f"Sorted list: {sorted_S}")
```
